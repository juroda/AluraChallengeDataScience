### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Calculo el ingreso total por tienda
ingreso_tienda = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

print(f"Ingreso total Tienda 1: {ingreso_tienda}")
print(f"Ingreso total Tienda 2: {ingreso_tienda2}")
print(f"Ingreso total Tienda 3: {ingreso_tienda3}")
print(f"Ingreso total Tienda 4: {ingreso_tienda4}")
```

```python
import matplotlib.pyplot as plt

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [ingreso_tienda, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]

plt.figure(figsize=(8,5))
plt.bar(tiendas, ingresos, color='mediumseagreen')
plt.title('Ingresos Totales por Tienda')
plt.ylabel('Ingresos ($)')
plt.xlabel('Tienda')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
# Agrupo los datos por categoría y cuento la cantidad de ventas por categoría en cada tienda
ventas_categoria_tienda = tienda.groupby('Categoría del Producto').size()
ventas_categoria_tienda2 = tienda2.groupby('Categoría del Producto').size()
ventas_categoria_tienda3 = tienda3.groupby('Categoría del Producto').size()
ventas_categoria_tienda4 = tienda4.groupby('Categoría del Producto').size()

print("Ventas por Categoría - Tienda 1")
print(ventas_categoria_tienda)
print("\nVentas por Categoría - Tienda 2")
print(ventas_categoria_tienda2)
print("\nVentas por Categoría - Tienda 3")
print(ventas_categoria_tienda3)
print("\nVentas por Categoría - Tienda 4")
print(ventas_categoria_tienda4)
```

```python
ventas_categoria_tienda.plot.pie(
    autopct='%1.1f%%',
    figsize=(8,8),
    startangle=90,
    shadow=True,
    explode=[0.05]*len(ventas_categoria_tienda),
    colors=plt.cm.Paired.colors
)

plt.title('Distribución de Ventas por Categoría - Tienda 1')
plt.ylabel('')  # Elimina etiqueta del eje Y
plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
# Caulo el promedio de cada tienda
calificacion_promedio_tienda1 = tienda['Calificación'].mean()
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()

print(f'Calificación promedio - Tienda 1: {calificacion_promedio_tienda1:.2f}')
print(f'Calificación promedio - Tienda 2: {calificacion_promedio_tienda2:.2f}')
print(f'Calificación promedio - Tienda 3: {calificacion_promedio_tienda3:.2f}')
print(f'Calificación promedio - Tienda 4: {calificacion_promedio_tienda4:.2f}')
```

# 4. Productos más y menos vendidos

```python
# TIENDA 1: Cuento las ventas por producto Tienda
ventas_productos_tienda1 = tienda['Producto'].value_counts()

producto_mas_vendido_t1 = ventas_productos_tienda1.idxmax()
ventas_mas_t1 = ventas_productos_tienda1.max()
producto_menos_vendido_t1 = ventas_productos_tienda1.idxmin()
ventas_menos_t1 = ventas_productos_tienda1.min()

print(f"Tienda 1 - Producto más vendido: {producto_mas_vendido_t1} ({ventas_mas_t1} ventas)")
print(f"Tienda 1 - Producto menos vendido: {producto_menos_vendido_t1} ({ventas_menos_t1} ventas)")

# TIENDA 2: Cuento las ventas por producto Tienda
ventas_productos_tienda2 = tienda2['Producto'].value_counts()
producto_mas_vendido_t2 = ventas_productos_tienda2.idxmax()
ventas_mas_t2 = ventas_productos_tienda2.max()
producto_menos_vendido_t2 = ventas_productos_tienda2.idxmin()
ventas_menos_t2 = ventas_productos_tienda2.min()

print(f"Tienda 2 - Producto más vendido: {producto_mas_vendido_t2} ({ventas_mas_t2} ventas)")
print(f"Tienda 2 - Producto menos vendido: {producto_menos_vendido_t2} ({ventas_menos_t2} ventas)")

# TIENDA 3
ventas_productos_tienda3 = tienda3['Producto'].value_counts()
producto_mas_vendido_t3 = ventas_productos_tienda3.idxmax()
ventas_mas_t3 = ventas_productos_tienda3.max()
producto_menos_vendido_t3 = ventas_productos_tienda3.idxmin()
ventas_menos_t3 = ventas_productos_tienda3.min()

print(f"Tienda 3 - Producto más vendido: {producto_mas_vendido_t3} ({ventas_mas_t3} ventas)")
print(f"Tienda 3 - Producto menos vendido: {producto_menos_vendido_t3} ({ventas_menos_t3} ventas)")


# TIENDA 4
ventas_productos_tienda4 = tienda4['Producto'].value_counts()
producto_mas_vendido_t4 = ventas_productos_tienda4.idxmax()
ventas_mas_t4 = ventas_productos_tienda4.max()
producto_menos_vendido_t4 = ventas_productos_tienda4.idxmin()
ventas_menos_t4 = ventas_productos_tienda4.min()

print(f"Tienda 4 - Producto más vendido: {producto_mas_vendido_t4} ({ventas_mas_t4} ventas)")
print(f"Tienda 4 - Producto menos vendido: {producto_menos_vendido_t4} ({ventas_menos_t4} ventas)")

```

# 5. Envío promedio por tienda

```python
# Costo
envio_promedio_tienda1 = tienda['Costo de envío'].mean()
envio_promedio_tienda2 = tienda2['Costo de envío'].mean()
envio_promedio_tienda3 = tienda3['Costo de envío'].mean()
envio_promedio_tienda4 = tienda4['Costo de envío'].mean()

print(f"Envío promedio Tienda 1: ${envio_promedio_tienda1:.2f}")
print(f"Envío promedio Tienda 2: ${envio_promedio_tienda2:.2f}")
print(f"Envío promedio Tienda 3: ${envio_promedio_tienda3:.2f}")
print(f"Envío promedio Tienda 4: ${envio_promedio_tienda4:.2f}")
```

```python
envios = [envio_promedio_tienda1, envio_promedio_tienda2, envio_promedio_tienda3, envio_promedio_tienda4]

plt.figure(figsize=(8,5))
plt.barh(tiendas, envios, color='coral')
plt.title('Costo de Envío Promedio por Tienda')
plt.xlabel('Valor Promedio del Envío ($)')
plt.ylabel('Tienda')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
